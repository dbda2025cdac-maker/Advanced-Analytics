```python
import pandas as pd
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('../datasets/Men ODI Team Match Results - 20th Century.csv')
df2 = pd.read_csv('../datasets/Men ODI Team Match Results - 21st Century.csv')

df = pd.concat([df1,df2])
df.head()
```

```python
#Historic Win rate of india
win_match_india = len(df[(df['Match'].str.contains("India")) & (df['Result'] == "Won")])
total_matches_india = len(df[df['Match'].str.contains("India")])
win_rate = win_match_india/total_matches_india
print(f"{ win_rate:.2f}")
```

```python
#Probability that india will win exactly 35 out of 60 matches
ans_a = binom.pmf(35,60,win_rate)
print(f"{ans_a:.2f}")
```

```python
#Probability that india wins more than 40 matches
# ans_b = binom.cdf(60,60,win_rate) - binom.cdf(40,60,win_rate)
ans_b = 1 - binom.cdf(40,60,win_rate)
print(ans_b)
```

```python
#Probability that india wins atleast 45 matches

ans_c = 1 - binom.cdf(44,60,win_rate)
print(ans_c)
```

```python
#Number of wins india will have in 60 matches

x = np.arange(0,61)
# x
pmf_values = binom.pmf(x,60,win_rate)
# print(pmf_values)
plt.figure(figsize = (12,6))
plt.plot(x,pmf_values)
plt.show()
# pmf_values
```
