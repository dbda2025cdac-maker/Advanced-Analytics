```python
import pandas as pd
import numpy as np
```

```python
df_2015 = pd.read_csv("../datasets/marathon_results_2015.csv")
df_2016 = pd.read_csv("../datasets/marathon_results_2016.csv")
df_2017 = pd.read_csv("../datasets/marathon_results_2017.csv")
df = pd.concat([df_2015, df_2016, df_2017])
original_row_count = len(df)
df.head()
```

Systematic sampling

```python
#::7 means that from start to end , pick every 7th record and [:] means all colums
sample_df = df.iloc[::7, :]
sample_row_count = len(sample_df)
```

```python
print(sample_df.head())
```

```python
print(original_row_count)
print(sample_row_count)
```

Stratified sampling

```python
total_males = len(df[df['M/F'] == 'M'])
total_females = len(df[df['M/F'] == 'F'])

sample_size = int(0.1 * original_row_count)
sample_males = int(0.1 * total_males)
sample_females = int(0.1 * total_females)
print(sample_size, sample_males, sample_females)
```

```python
#Generate proportions samples of males and females atheletes
male_sample = df[df['M/F'] == 'M'].sample(n=sample_males, random_state = 42)
female_sample = df[df['M/F'] == 'F'].sample(n=sample_females, random_state = 42)
print(len(male_sample), len(female_sample))
```

```python
#Combine the male and female samples into a single sample dataframe

sample_df = pd.concat([male_sample,female_sample])
sample_row_count = len(sample_df)
print(df.head())
print(sample_df.head())
```

```python
print(original_row_count)
print(sample_row_count)
```

```python
print(f"Number of male atheletes in the sample dataframe : {len(male_sample)}")
print(f"Number of female atheletes in the sample dataframe : {len(female_sample)}")
```

# Cluster Sampling

```python
#Cluster Sampling

unique_countries = df['Country'].unique()
num_countries = len(unique_countries)
print(unique_countries, num_countries)
```

```python
#Randomly select half of the countires

np.random.seed(42)
#choice() function in numpy generates a random sample from an array
selected_countries = np.random.choice(unique_countries, size = num_countries//2, replace = False)
#Replace false means that do not allow same country twice
print(selected_countries)
```

```python
#Create a new df containing a;; recprds pf the selected countries
sample_df = df[df['Country'].isin(selected_countries)]
sample_row_count = len(sample_df)
print(df.head(), sample_df.head())
print(original_row_count, sample_row_count)
```
