```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/tips.csv')
df.head()
```

```python
#outlier analysis on total bill column of tips

data_array = df['total_bill']

q1 = np.percentile(df['total_bill'], 25)
q2 = median = np.median(df['total_bill'])
q3 = np.percentile(df['total_bill'], 75)

iqr = q3 - q1

lower_fence = q1 - (1.5 * iqr)
upper_fence = q3 + (1.5 * iqr)

outliers  = data_array[(data_array < lower_fence) | (data_array > upper_fence)]
```

```python
print(f"Q1 : {q1}")
print(f"Q2 : {q2}")
print(f"Q3 : {q3}")
print(f"iqr : {iqr}")
print(f"lower_fence : {lower_fence}")
print(f"upper_fence : {upper_fence}")
print(f"outliers : {outliers}")
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(data = data_array, orient = 'h')
plt.title('Box plot')
plt.xlabel('Bill amount')
plt.show()
```
