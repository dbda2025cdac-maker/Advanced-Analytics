```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/airline_passenger_satisfaction.csv')
onBoardService = df['On-board Service']
df.head()
```

```python
mean_onBoardService = onBoardService.mean()
median_onBoardService = onBoardService.median()
skew_onBoardService = onBoardService.skew()
kurt_onBoardService = onBoardService.kurt()
# p2 = onBoardService.kurtosis()

print(f"mean_onBoardService is : {mean_onBoardService}")
print(f"median_onBoardService is : {median_onBoardService}")
print(f"skew_onBoardService is : {skew_onBoardService}")
print(f"kurt_onBoardService is : {kurt_onBoardService}")
```

```python
plt.figure(figsize = (8,6))
sns.histplot(onBoardService,kde =True, color = 'skyblue' )
plt.axvline(mean_onBoardService, color = 'red', linestyle = '--', label = f"mean_onBoardService: {mean_onBoardService}")
plt.axvline(median_onBoardService, color = 'red', linestyle = '--', label = f"median_onBoardService: {median_onBoardService}")
plt.title(f"frequency distribution of final worth\n skewness : {skew_onBoardService:.2f}, kurtosis : {kurt_onBoardService:.2f}")
plt.xlabel('distribution of onBoardService')
plt.ylabel('Frequncy')
plt.tight_layout()
plt.show()
```

```python
#Compare two CoV Corefficinet of varience for two categories

group1_name = 'First-time'
group2_name = 'Returning'

group1 = df[df['Customer Type'] == group1_name]['Ease of Online Booking']
group2 = df[df['Customer Type'] == group2_name]['Ease of Online Booking']

#Calculate CoV 
cov_group1 = group1.std()/group1.mean() * 100
cov_group2 = group2.std()/group2.mean() * 100


print(f"Coefficient of Variance (CoV) for {group1_name} : {cov_group1:.2f}%")
print(f"Coefficient of Variance (CoV) for {group2_name} : {cov_group2:.2f}%")
```
