# One Sample t test

```python
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv('../datasets/Mall_customers.csv')
age_data = data['Age']
```

```python
age_data = age_data.sample(n=30, replace = False)
pop_avg_age = 40 # H0 Mall customers have an average age of 40

t_statistic,p_value = stats.ttest_1samp(age_data, pop_avg_age)
print("P-Value: ", p_value)
```

```python
alpha = 0.05
if p_value > alpha:
    print("Do not reject Ho")
else:
    print("Reject H0")
```
