```python
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df['age_mean'] = df['age'].fillna(df['age'].mean())
df['age_median'] = df['age'].fillna(df['age'].median())
df['age_mode'] = df['age'].fillna(df['age'].mode()[0])
df['age_constant'] = df['age'].fillna(0)
```

```python
df
```

```python
#handling missing values with simpleimputed mean of scikitlearn library

imp_mean = SimpleImputer(strategy='mean')
df['age_imputed_mean'] = imp_mean.fit_transform(df[['age']])

imp_median = SimpleImputer(strategy='median')
df['age_imputed_median'] = imp_median.fit_transform(df[['age']])

imp_mode = SimpleImputer(strategy='most_frequent')
df['age_imputed_mode'] = imp_mode.fit_transform(df[['age']])

imp_costant = SimpleImputer(strategy = 'constant', fill_value = 0)
df['age_imp_const'] = imp_costant.fit_transform(df[['age']])

df
```
