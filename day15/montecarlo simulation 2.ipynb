# Startup investment decision: MonteCarlo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# print(stock_returns)
# print(bond_returns)
# total_returns_pct = ((stock_weight * stock_returns) + (bond_weight * bond_returns)) * 100

#Parameters
initial_investment = 100000 #investment amount in dollars
success_rate = 0.01
failure_rate = 0.5
success_multiplier = 1000000
num_simulations = 10000
```

```python
#Monte Carlo Simulaiton
final_values = []
for _ in range(num_simulations):
    #Simulations whether the startup succeeds (1 if success, 0 otherwise)
    success = np.random.choice([1,0],p = [success_rate, 1 - success_rate])
    if success :
        final_value = initial_investment * success_multiplier
    else:
        final_value = initial_investment * (1-failure_rate)
    final_values.append(final_value)
```

```python
#Results
mean_final_value = np.mean(final_values)
risk_of_loss = np.mean(np.array(final_values) < (initial_investment))
```

```python
print(f"Mean Final Values: ${mean_final_value:.2f}")
print(f"Probability of Loss: {risk_of_loss * 100:.2f}%")
```

```python
#Plotting the distribution
plt.hist(final_values, bins = 30, edgecolor = 'k')
plt.title('Simulated Startups Investment Outcomes')
plt.xlabel('Investment Value ($)')
plt.ylabel('Frequency')
plt.show()
```
