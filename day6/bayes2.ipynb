```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/loan_approval_dataset.csv')
N = len(df)
A = len(df[df['loan_status'] == 'Approved'])
B = len(df[df['cibil_score'] >= 700])
```

```python
A_and_B = len(df[(df['loan_status'] == 'Approved') & (df['cibil_score']>=700)])
#A_and B  = approved and good credit
```

```python
#probabilities
P_A = A/N                   #P(loan approved)
P_B = B/N                   #P(Good credit)
P_B_given_A = A_and_B/A     #P(Good credit | approved)
```

```python
#Bayes theorem
P_A_given_B = (P_B_given_A * P_A)/P_B
```

```python
print("Total records: ", N)
print("Approved count", A)
print("Good credit  count", B)
print("Approved and good credit: ", A_and_B)
print("P(Loan Approved):", P_A)
print("P(Good cibil score):", P_B)
print("P(Good cibil score | Loan was Approved):", P_B_given_A)
print("P(Loan Approved | good cibil score):", P_A_given_B)
```

```python
#Two new customers walk in, one with cibil score 800 AND ANOTHER WITH 500
#bAD CREDIT COUNT
bad_Credit = N-B #People with cibil score < 700

#Aproved and bad credit
A_and_notB = A - A_and_B

#P(Loan approved and bad credit)
P_A_given_notB = (A_and_notB/A)*(P_A / (bad_Credit/N))
```
