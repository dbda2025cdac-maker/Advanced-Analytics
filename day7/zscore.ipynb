```python
 #Z-score caluclaiton, outlier detetion and visualization
import pandas as pd
from scipy.stats import zscore
```

```python
df = pd.read_csv('../datasets/weight-height.csv')

weight_mean = df['Weight'].mean()
weight_Std = df['Weight'].std()

height_mean = df['Height'].mean()
height_Std = df['Height'].std()
# df.head()
```

```python
df['Weight_zscore'] = zscore(df['Weight'])
df['Height_zscore'] = zscore(df['Height'])
```

```python
print(f"Mean of Weight: {weight_mean:.2f}")
print(f"STD of Weight: {weight_Std:.2f}")

print(f"Mean of Height: {height_mean:.2f}")
print(f"STD of Height: {height_Std:.2f}")
```

```python
#Outlier analysis
zscore_threshold = 3
```

```python
weight_outliers = df[df['Weight_zscore'].abs() > zscore_threshold]
height_outliers = df[df['Height_zscore'].abs() > zscore_threshold]

all_outliers = pd.concat([weight_outliers, height_outliers]).drop_duplicates()
```

```python
print(f"\nNumber of Weight Outliers (zscore > {zscore_threshold}): {len(weight_outliers)}")
print(f"\nNumber of Height Outliers (zscore > {zscore_threshold}): {len(height_outliers)}")
print(f"\nTotal unique outliers found : {len(all_outliers)}")
```

```python
#Plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (14,6))

#Plot for weight
plt.subplot(1,2,1) #1row, 2 columns, 1st plot
sns.histplot(df['Weight'], kde=True, color = 'skyblue')
# plt outliers at y=0 for clarity
plt.scatter(weight_outliers['Weight'], [0]*len(weight_outliers), color ='red',s=50, zorder = 5,label = f"Outliers (z > {zscore_threshold})")
plt.axvline(weight_mean, color = 'green', linestyle = '--', label = f"Mean : {weight_mean:.2f}")
plt.title("Distribution of Wieght with Outliers")
plt.xlabel('Weight (lbs)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.7)

#Plot for hieght
plt.subplot(1,2,2) #1row 2 columns and 2nd plot
sns.histplot(df['Height'], kde=True, color = 'orange')
# plt outliers at y=0 for clarity
plt.scatter(height_outliers['Height'], [0]*len(height_outliers), color ='red',s=50, zorder = 5,label = f"Outliers (z > {zscore_threshold})")
plt.axvline(height_mean, color = 'green', linestyle = '--', label = f"Mean : {height_mean:.2f}")
plt.title("Distribution of Hieght with Outliers")
plt.xlabel('Height (lbs)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.7)

plt.tight_layout()
plt.show()
```

```python
#Scatte plot for Height vs Weight with outliers

plt.figure(figsize = (10,7))
sns.scatterplot(x='Weight', y='Height', data = df, color = 'blue', alpha = 0.6, label = 'Normal Data')
sns.scatterplot(x='Weight', y='Height', data = all_outliers, color = 'red', s=100, label = 'Outliers', marker = 'X')
plt.title('Height vs Weight Outliers highlighted')
plt.xlabel('Weight(lbs)')
plt.ylabel('Height (inches)')
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.7)
plt.show()
```

```python
#Box plot for weight
plt.figure(figsize=(6,8))
sns.boxplot(y=df['Weight'], color = 'skyblue')
plt.title('Box plot of weight')
plt.ylabel('Weight (lbs)')
plt.grid(True, linestyle = '--', alpha = 0.7)
plt.show()
```

```python
#Box plot for height
plt.figure(figsize=(6,8))
sns.boxplot(y=df['Height'], color = 'skyblue')
plt.title('Box plot of height')
plt.ylabel('Height (lbs)')
plt.grid(True, linestyle = '--', alpha = 0.7)
plt.show()
```
