```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/titanic.csv')
```

```python
df.info()
```

```python
total_passangers = len(df)
survived_count = len(df[df['survived'] == 1])
female_count = len(df[df['gender'] == 'female'])
# female_survived_count = len(df[(df['gender'] == 'female') & (df[df['survived'] == 1])])
female_survived_count = len(df[(df['gender'] == 'female') & (df['survived'] == 1)])
```

```python
P_survived  = survived_count/total_passangers
P_female = female_count/total_passangers
P_female_given_survived = female_survived_count/survived_count
```

```python
P_survived_give_female = (P_female_given_survived * P_survived)/P_female
print(f"Probablilty of survival given that passangers is female : {P_survived_give_female}")
```
