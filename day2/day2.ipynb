```python
import numpy as np
import pandas as pd
```

```python
a = np.array([10,20,30])
b = np.array([1,2,3])
print(a+b)
```

```python
s1 = pd.Series([10,20,30,40], index = ['x','y','z','a'])
s2 = pd.Series([1,2,3], index = ['y','z','x'])
print(s1+s2)
```

Creating a PD Series

```python
# M-1
names = ['Alice', 'Bob', 'Charlie']
marks = [67,31,81]

stu = pd.Series(data = marks)
print(stu)

stu = pd.Series(data = marks, index = names)
print(stu)
```

```python
# M-2
random_ages = np.random.randint(0,100,4)
names.append('David')
people_arr = pd.Series(data = random_ages, index = names)
print(people_arr)
```

```python
# M-3
names_ages_dict = {'Alice':21, 'Bob':25, 'Charlie':19, 'David':23}
print(pd.Series(names_ages_dict))
```

```python
#Creting imaginary sales data for quater 1 and quater 2
q1 = {'Japan':80, 'China':450, 'India':200, 'USA':250, 'Mexico':150}
q2 = {'Japan':100, 'China':550, 'India':210, 'USA':260}

sales_q1 = pd.Series(q1)
sales_q2 = pd.Series(q2)

print(sales_q1)
print(sales_q1['Japan'])
print(sales_q2[0])
```

```python
print(sales_q1.keys())
print(sales_q1*2)
print(sales_q2 / 100)
```

```python
#Handling missing values in pd

print(sales_q1 + sales_q2)
sales_q1.add(sales_q2, fill_value=0)
```

```python
#Creating a pd dataframe

scores = np.random.randint(0,101, (4,3)) #4players 3 matches
print(scores)

players = ['Virat', 'Rohit', 'Dhoni', 'Jadeja'] #index
matches = ['match1', 'match2', 'match3'] #3columns

df = pd.DataFrame(data=scores)
print(df)
```

```python
df = pd.DataFrame(data = scores, index = players)
print(df)
```

```python
df = pd.DataFrame(data = scores, index = players, columns = matches)
print(df)
```
