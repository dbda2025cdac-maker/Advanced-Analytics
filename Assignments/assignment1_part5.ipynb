```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/social_media_usage.csv')
df.head()
```

```python
df_facebook = df[df['App'] == 'Facebook']
df_instagram = df[df['App'] == 'Instagram']

print(df_facebook)
print(df_instagram)
```

```python
cov_overall = df[['Daily_Minutes_Spent','Posts_Per_Day']].cov()
cov_facebook = df_facebook[['Daily_Minutes_Spent','Posts_Per_Day']].cov()
cov_instagram =df_instagram[['Daily_Minutes_Spent','Posts_Per_Day']].cov()

print("Covariance is :")
print(cov_overall)
print(cov_facebook)
print(cov_instagram)
```

```python
corr_overall = df[['Daily_Minutes_Spent','Posts_Per_Day']].corr()
corr_facebook = df_facebook[['Daily_Minutes_Spent','Posts_Per_Day']].corr()
corr_instagram =df_instagram[['Daily_Minutes_Spent','Posts_Per_Day']].corr()

print("Corelation is :")
print(corr_overall)
print(cov_facebook)
print(corr_instagram)
```

```python
fig,axes = plt.subplots(1,3,figsize = (18,6))
axes[0].scatter(df['Daily_Minutes_Spent'],df['Posts_Per_Day'], color = 'green', alpha = 0.5)
axes[0].set_title(f"Overall correlation : {corr_overall.loc['Daily_Minutes_Spent','Posts_Per_Day']:.2f}")
axes[0].set_xlabel("Daily_Minutes_Spent")
axes[0].set_ylabel("Posts_Per_Day")
axes[0].grid(True)

axes[1].scatter(df_facebook['Daily_Minutes_Spent'],df_facebook['Posts_Per_Day'], color = 'blue', alpha = 0.5)
axes[1].set_title(f"Overall correlation : {corr_facebook.loc['Daily_Minutes_Spent','Posts_Per_Day']:.2f}")
axes[1].set_xlabel("Daily_Minutes_Spent")
axes[1].set_ylabel("Posts_Per_Day")
axes[1].grid(True)

axes[2].scatter(df_instagram['Daily_Minutes_Spent'],df_instagram['Posts_Per_Day'], color = 'pink', alpha = 0.5)
axes[2].set_title(f"Overall correlation : {corr_instagram.loc['Daily_Minutes_Spent','Posts_Per_Day']:.2f}")
axes[2].set_xlabel("Daily_Minutes_Spent")
axes[2].set_ylabel("Posts_Per_Day")
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
#Heatmap
plt.figure(figsize = (18,6))
plt.subplot(1,2,1)
sns.heatmap(corr_facebook, annot = True, fmt = '.2f', cmap = "coolwarm")
plt.title('Facebook Correlation Heatmap')
plt.xticks(rotation = 0)
plt.yticks(rotation = 0)
# plt.tight_layout()
# plt.show()


plt.subplot(1,2,2)
sns.heatmap(corr_instagram, annot = True, fmt = '.2f', cmap = "coolwarm")
plt.title('Instagram Correlation Heatmap')
plt.xticks(rotation = 0)
plt.yticks(rotation = 0)
plt.tight_layout()
plt.show()
```
