```python
#IQR outlier analysis of computing times with boxplot

import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
commuter_times = [16, 8,35,17,13,15,15,5,16,25,20,20,12,10]
data_array = np.array(commuter_times)
```

```python
q1 = np.percentile(data_array, 25) #25th percentile = Q1
q3 = np.percentile(data_array, 75) #25th percentile = Q3

iqr = q3 - q1

lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

outliers  = data_array[(data_array < lower_fence) | (data_array > upper_fence)]
```

```python
commuter_times.sort()
print(commuter_times)
print(f"Q1 : {q1}")
print(f"Q2 : {np.median(data_array)}")
print(f"Q3 : {q3}")
print(f"iqr : {iqr}")
print(f"lower_fence : {lower_fence}")
print(f"upper_fence : {upper_fence}")
print(f"outliers : {outliers}")
```

```python
#Create the boxplot
plt.figure(figsize = (10,6))
sns.boxplot(data = commuter_times, orient = 'h')
plt.title('Box Plot of Commuter Times')
plt.xlabel('Minutes')
plt.show()
```
