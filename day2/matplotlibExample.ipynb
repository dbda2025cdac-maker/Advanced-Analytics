```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tips.csv')
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['total_bill'], bins = 10, color = 'skyblue' , edgecolor = 'black')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
#same plot using sns

plt.figure(figsize = (6,4))
sns.histplot(df['total_bill'],bins = 10, kde= True,color = 'salmon')
#kde stands for kjernal density function
plt.title('Distribution of total Bill with Seaborn')
plt.show()
```

```python
#bar plot of average tip by gender
avg_tip_by_gender = df.groupby('gender')['tip'].mean() 
#grouping automatically creates the index on the grouped columns, in this case gender
plt.figure(figsize= (6,4))
plt.bar(avg_tip_by_gender.index, avg_tip_by_gender.values, color = ['lightpink', 'lightblue'])
#here index = gender
plt.title('Average Tip by Gender')
plt.ylabel('Average Tip')
plt.show()
```

```python
#using seaborn

avg_tip_by_gender = df.groupby('gender')['tip'].mean().reset_index() 
#reset_index() to convert index back to column
plt.figure(figsize = (6,4))
sns.barplot(x = 'gender', y = 'tip', data = avg_tip_by_gender, palette=['lightblue','pink'])
plt.title('Average Tip by Gender')
plt.ylabel('Average Tip')
plt.show()
```

```python
#scatter plot of total bill vs tip - tells us how these are related to each other

plt.figure(figsize = (6,4))
plt.scatter(df['total_bill'], df['tip'], color = 'purple', edgecolor = 'black')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
#with seaborn, but also with regression line
plt.figure(figsize = (6,4))
sns.regplot(x = 'total_bill', y = 'tip', data = df, color = 'green')
plt.title('Total Bill vs Tip with Regression Line')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
