Consider the diabetes dataset and find the relation of age with diabetes outcome
group age in 3 categories young : <=30 middle 31-50 and old > 50

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np
```

```python
df = pd.read_csv('../datasets/diabetes.csv')
df.head()
```

```python
conditions = [
    (df['Age'] >= 0) & (df['Age'] <= 30), 
    (df['Age'] >= 31) & (df['Age'] <= 50),
    (df['Age'] > 50) 
]

# 2. Define your choices corresponding to the conditions
groups = ['young', 'middle', 'old']

# 3. Use np.select
# 'default' handles values that don't match any condition (e.g., negative ages or NaNs)
df['Age Group'] = np.select(conditions, groups, default='unknown')
```

```python
df
```

```python
contingency_table = pd.crosstab(df['Age Group'], df['Outcome'])
contingency_table
```

```python
chi2_stat, p_value,dof, expected  = chi2_contingency(contingency_table)
```

```python
print("Chi-Sq statistic : ", chi2_stat)
print("p_value: ", p_value)
print("degree of freedom : ", dof)
```

```python
if p_value < 0.05:
    print("Reject H0")
else:
    print("Do not reject H0")
```
