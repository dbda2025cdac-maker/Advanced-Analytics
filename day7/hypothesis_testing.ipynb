```python
import pandas as pd
import scipy.stats as stats
```

```python
df = pd.read_csv('../datasets/rainfall_new_jersery.csv')
df
```

```python
year = df.YearTotalmm
shapiro_test = stats.shapiro(year)
print(f"Test Statistics (W) :{shapiro_test.statistic}")
print(f"Pvalue: {shapiro_test.pvalue}")
```
