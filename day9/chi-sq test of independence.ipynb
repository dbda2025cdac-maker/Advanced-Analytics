# chi-sq test of independence, also first create the contingency table using crosstab() function 
# H0: there is no dependency of passangers class on passanger survival 

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv('../datasets/titanic.csv')
```

```python
df
```

```python
# create a congingency table
contingency_table = pd.crosstab(df['pclass'], df['survived'])
print(contingency_table)
```

```python
#perform the chi-sq test

chi2_stat, p_value ,dof,expected = chi2_contingency(contingency_table)
```

```python
print("Chi-Sq statistic : ", chi2_stat)
print("p_value: ", p_value)
print("degree of freedom : ", dof)
```

```python
if p_value < 0.05:
    print("Reject H0")
else:
    print("Do not reject H0")
```
