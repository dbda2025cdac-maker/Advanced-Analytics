```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('tips.csv')
```

```python
print(df.columns)
print(df.index)
print(df.head())
print(df.tail())
print(df.info())
```

```python
print(len(df))
print(df.describe())
print(df.describe().transpose())
```

```python
print(df['total_bill'])
print(df.total_bill) #another way for the same thing
print(df[['total_bill', 'tip']])
```

```python
#Add a  new column
df['tip_percentage'] = 100 * df['tip']/df['total_bill']
df['price_per_person'] = df['total_bill']/df['size']
print(df.head())
df['price_per_person'] = np.round(df['price_per_person'], 2)
print(df.head())
```

```python
# df = df.drop('tip_percentage', axis = 1) #axis = means drop a columns
#OR
df.drop('tip_percentage', axis = 1, inplace=True) #this will make the change permanent
print(df.head())
```

```python
#To set a column as an index
df = df.set_index('Payment ID')
print(df.head())
df = df.reset_index()
print(df.head())
```

```python
#coonditional filtering

print(df['total_bill'] < 30) #Boolean Series
print(df[df['total_bill'] < 30]) #actual rows where total_billl < 30
print(df[df['gender'] == 'Male'])
print(df[(df['total_bill'] > 30) & (df['gender'] == 'Male')])
print(df[(df['total_bill'] > 30) &  ~(df['gender'] == 'Male')])
print(df[(df['total_bill'] > 30) |  (df['gender'] == 'Male')])
```

```python
options = ['Sat', 'Sun']
print(df[df['day'].isin(options)])
```

```python
# Multiple syntax

filtered1 = df[(df['gender'] == 'Female') & (df['tip'] >= 4)]
filtered2 = df.query("gender == 'Female' and tip >= 4")
filtered3 = df.loc[(df['gender'] == 'Female') & (df['tip'] >= 4)]
filtered4 = df.where((df['gender'] == 'Female') & (df['tip'] >= 4))

print(filtered1)
print(filtered2)
print(filtered3)
print(filtered4)
```

```python
gender_count = df['gender'].value_counts()
print(gender_count)

sorted_by_bill = df.sort_values(by = 'total_bill').head(5)
print(sorted_by_bill)

sorted_by_tip_desc = df.sort_values(by='tip', ascending = False).head(5)
print(sorted_by_tip_desc)

sorted_multi = df.sort_values(by = ['day', 'total_bill'], ascending = [True, False]).head(10)
print(sorted_multi)
```
