```python
#LPP solution using PuLP
from pulp import *
```

```python
#create the problem variable to contain the problem data
model = LpProblem("FactoryProductionProblem",LpMaximize)
```

```python
#create variables for chairs and tables
#Parameters : name, Lower limit, upper limit , datatype
chairs = LpVariable("Chairs", 0,None, LpInteger) #Number of chairs to be produced
tables = LpVariable("Tables", 0,None, LpInteger) #Number of tables to be produced
```

```python
#Create maximize objective fucntion (profit)
model += 300 * chairs + 1000 * tables, "Profit"
```

```python
#Create constraints
model += 5 * chairs + 10 * tables <= 400, "Wood Cosntraint" #Wood availability constraint
model += 2 * chairs + 6 * tables <= 300, "Time Cosntraint" #Worker Time constraint
```

```python
#The problem is solved using PuLP's choice of Solver
model.solve()
```

```python
#Each of the variable is printed with its resolved optmum value
for v in model.variables():
    print(v.name, " = ", v.varValue)
```

```python
#Additional Information
print("Total Profit: ", value(model.objective)) #Print the objective function value (total profit)
```
