# Calculate zscores and do the outlier analysis on:
    1. Glucose column of Diabetes Datset
    2. Billionaires datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

```python
df = pd.read_csv('../datasets/diabetes.csv')
df.head()
```

```python
df['glucose_zscores'] = zscore(df['Glucose'])
zscore_threshold = 3
```

```python
glucose_mean = df['Glucose'].mean()

glucose_outliers = df[df['glucose_zscores'].abs() > zscore_threshold]
len(glucose_outliers)
```

```python
plt.figure(figsize = (14,6))
plt.subplot(1,2,1) #1row, 2 columns, 1st plot
sns.histplot(df['Glucose'], kde=True, color = 'skyblue')
# plt outliers at y=0 for clarity
plt.scatter(glucose_outliers['Glucose'], [0]*len(glucose_outliers), color ='red',s=50, zorder = 5,label = f"Outliers (z > {zscore_threshold})")
plt.axvline(glucose_mean, color = 'green', linestyle = '--', label = f"Mean : {glucose_mean:.2f}")
plt.title("Distribution of Gluscose with Outliers")
plt.xlabel('Glucose')
# plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.7)
```
