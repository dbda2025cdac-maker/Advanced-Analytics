```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
df = pd.read_csv('../datasets/tips.csv')
print("Original Dy columns :\n", df['day'].head())
```

```python
#One Hot Encoding

ohe = OneHotEncoder(sparse_output=False, drop = None)
#Drop none = keep all columns
# sparse_output = False -> Compress columns ie store only 1s and not 0s

ohe_array = ohe.fit_transform(df[['day']])
```

```python
#Column names created by one hot encoder
ohe_columns = ohe.get_feature_names_out(['day'])

#Create new dataframe with one hot endoer columns
df_ohe = df.join(pd.DataFrame(ohe_array, columns=ohe_columns, index = df.index))
```

```python
print('\nOne Hot Encoded "day": \n', df_ohe[ohe_columns].head())
print('\nOne Hot Encoded "day": \n', df_ohe[ohe_columns].tail())
```

```python
# ------------Drop one redundand categories----------
ohe_drop = OneHotEncoder(sparse_output=False,drop = 'first')
#drop = first drops one category

ohe_array2 = ohe_drop.fit_transform(df[['day']])
ohe_cols2 = ohe_drop.get_feature_names_out(['day'])

df_ohe2 = df.join(pd.DataFrame(ohe_array2,columns = ohe_cols2, index = df.index))
print('\nOne Hot Encoded "day": \n', df_ohe2[ohe_cols2].head())
print('\nOne Hot Encoded "day": \n', df_ohe2[ohe_cols2].tail())
```

# Label Encoding

```python
label_encoder = LabelEncoder()
df['day_label'] = label_encoder.fit_transform(df['day'])
print('\n Label Encoded "day" :\n ', df[['day', 'day_label']].head())
print('\n Label Encoded "day" :\n ', df[['day', 'day_label']].tail())
```

# Frequency Encoding

```python
day_freq = df['day'].value_counts(normalize=False)
#normalize = False means use %
#normalize = False means use count instead 
df['day_freq'] = df['day'].map(day_freq)
print("\nFrequency Encoded 'day' :\n", df[['day','day_freq']].head())
print("\nFrequency Encoded 'day' :\n", df[['day','day_freq']].tail())
```
