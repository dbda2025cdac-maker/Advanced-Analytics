```python
#4th part
avg_age_male = df['age'].where(df['gender'] == 'male').mean()
avg_age_female = df['age'].where(df['gender'] == 'female').mean()

print(f"Average male age is {avg_age_male:.0f}")
print(f"Average female age is {avg_age_female:.0f}")
```

```python
#5th part

# df.groupby('pclass').count()
print(df.value_counts('pclass'))
df['pclass'].value_counts()
```

```python
#6th part

survived_df = df[df['survived'] == 1]
not_survived_df = df[df['survived'] == 0]

print(survived_df)
print(not_survived_df)
```

```python
#7th part


pd.set_option('display.max_rows',None)

females_survived = survived_df[survived_df['gender'] == 'female']['gender'].count()
males_survived = survived_df[survived_df['gender'] == 'male']['gender'].count()

females_not_survived = not_survived_df[not_survived_df['gender'] == 'female']['gender'].count()
males_not_survived = not_survived_df[not_survived_df['gender'] == 'male']['gender'].count()

print(f'females_survived : {females_survived}')
print(f'males_survived : {males_survived}')
print(f'females_not_survived : {females_not_survived}')
print(f'males_not_survived : {males_not_survived}')

pd.reset_option('display.max_rows')
```
