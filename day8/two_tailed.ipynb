'''
Consider H0 = doing pg-dbda course does not import the chances of placements and they remain 40%

```python
import math
from scipy.stats import norm
```

```python
#Given
mu = 40
sample_means = [80, 41,45, -35]
n=100
s=15
alpha = 0.5
```

```python
#Critical values
#norm.cdf() gives probability, norm.pdf() gives zscore 
# ppf  = percentage point function (Inverse of CDF)
# gives z score corresponing to a giveen probability
#e..g tell me the z score such that the aqrea inder the curve (left of it) is p
z_crit_two_tailed = norm.ppf(1-alpha/2) #(±)1.96
z_crit_one_tailed = norm.ppf(1-alpha) #+1.645

print("critical values:")
print('=================')
print(f'Two-Tailed alpha = (0.05) :±{z_crit_two_tailed:.3f}')
print(f'One-Tailed alpha = (0.05) :{z_crit_one_tailed:.3f}\n')
```

```python
#Function to compute Z
def compute_z(xbar):
    return (xbar - mu)/(s/math.sqrt(n))
```

```python
#evaluation
for xbar in sample_means:
    print(f'=====For sample_mean x bar = {xbar}=====')

    z = compute_z(xbar)

    #Pvalues
    #norm.cdf(z) give the probability to the LEFT of z
    #So 1- norm.cdf(z) gives the probability to the RIGHT of Z
    #For two tailed test: compute to the right and double
    #eg for z=2 right tailed area = 0.0228
    # So total area = 2 * 0.0228 = 0.0456
    # For one tailed one sided only 
    p_two_tailed = 2 * (1-norm.cdf(abs(z)))
    p_one_tailed = 1 - norm.cdf(z) # imporvement only (right-tail)

    print(f"Z-score : {z:.3f}")
```

```python
#Two tailed decision
if abs(z) > z_crit_two_tailed :
    two_tail_decision = "Reject H0 (significant difference)"
else:
    two_tail_decision = "Fail to reject H0 (No significant difference)"
```

```python
#One Tailed Decision
if z > z_crit_one_tailed:
    one_tail_decision = "Reject H0 (significant imporvement)"
else:
    one_tail_decision = "Fail to reject H0 (No significant  improvement)"
```

```python
print(f"Two tailed test:")
print(f" p-value : {p_two_tailed:.4f}")
print(f"Decision :{two_tail_decision}")
```

```python
print(f"One tailed test (improvement only)")
print(f" p-value = {p_one_tailed:.4f}")
print(f" Decision : {one_tail_decision}\n")
```
