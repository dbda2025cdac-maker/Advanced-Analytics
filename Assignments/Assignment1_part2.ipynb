```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/titanic.csv')
df
```

```python
df = df[df['age'].notna()]
```

```python
#outlier analysis on age column of titanic dataset

data_array = df['age']

q1 = np.percentile(data_array, 25)
q2 = median = np.median(data_array)
q3 = np.percentile(data_array, 75)

iqr = q3 - q1

lower_fence = q1 - (1.5 * iqr)
upper_fence = q3 + (1.5 * iqr)

outliers  = data_array[(data_array < lower_fence) | (data_array > upper_fence)]
```

```python
print(f"Q1 : {q1}")
print(f"Q2 : {q2}")
print(f"Q3 : {q3}")
print(f"iqr : {iqr}")
print(f"lower_fence : {lower_fence}")
print(f"upper_fence : {upper_fence}")
print(f"outliers : {outliers}")
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(data = data_array, orient = 'h')
plt.title('Box plot')
plt.xlabel('age')
plt.show()
```
