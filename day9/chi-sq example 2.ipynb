Find if gender and smoking is related in tips dataset

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv('../datasets/tips.csv')
```

```python
df
```

```python
contingency_table = pd.crosstab(df['gender'], df['smoker'])
contingency_table
```

```python
chi2_stat, p_value,dof, expected  = chi2_contingency(contingency_table)
```

```python
print("Chi-Sq statistic : ", chi2_stat)
print("p_value: ", p_value)
print("degree of freedom : ", dof)
```

```python
if p_value < 0.05:
    print("Reject H0")
else:
    print("Do not reject H0")
```
